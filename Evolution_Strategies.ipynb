$x^*=\underset{x}{\arg\min}f(x)=\underset{x}{\arg\min}~x^2$

```python
def f1(x):
    return x**2
```

```python
x = np.linspace(-3,3,100)
plt.figure()
plt.plot(x, f1(x))
plt.xlabel('x', fontsize=15)
plt.ylabel('f(x)', fontsize=15)
```

$x^*=\underset{x}{\arg\min}~f(x)=\underset{x}{\arg\min}~x^4+2.5x^3-0.5x^2-x+1$

```python
def f2(x):
    return x**4 + 2.5*(x**3) - 0.5*(x**2) - x + 1
```

```python
x = np.linspace(-3,3,100)
plt.figure()
ax = plt.gca()
ax.set_aspect(0.5)
plt.plot(x, f2(x))
plt.xlabel('x', fontsize=15)
plt.ylabel('f(x)', fontsize=15)
plt.ylim(-4,4)
plt.xlim(-4,2)
```

```python
def gradient_f2(x):
    return 4.0*(x**3) + 3.0*2.5*(x**2) - x - 1.0
```

```python
def gd(fobj, gradient, x_init, learning_rate, max_iters=50):
    all = []
    x = x_init
    iter = 0
    while True:
        value = fobj(x)
        grad = gradient(x)
        all.append((x, value, grad))
        x = x - learning_rate*grad
        iter += 1
        if np.linalg.norm(np.fabs(grad)) < 1e-5:
            break
        if iter > max_iters:
            break
    return all
```

```python
all = gd(f2, gradient_f2, 1.5, 0.01, 50)
```

```python
x = np.linspace(-3,3,100)

for iteration in range(0, len(all)):
    plt.figure()
    ax = plt.gca()
    ax.set_aspect(0.5)
    plt.plot(x, f2(x))
    plt.xlabel('x', fontsize=15)
    plt.ylabel('f(x)', fontsize=15)
    plt.ylim(-4,4)
    plt.xlim(-4,2)

    plt.scatter(all[iteration][0], all[iteration][1], c='red', s=50)
    plt.show()
    time.sleep(0.1)
    if iteration < len(all):
        display.clear_output(wait=True)
```

```python
def Peak(x, y):
    return 3.0*(1.0 - x)**2*np.exp(-x**2-(y+1.0)**2) - 10*(x/5.0-x**3-y**5)*np.exp(-x**2-y**2)-1.0/3.0*np.exp(-(x+1.0)**2-y**2)
```

```python
def es(fobj, bounds, sigma_init=1.0, c_inc=1.1, c_dec=0.6, lambd=20, max_iters=100):
    dimensions = len(bounds)

    lower_bound, upper_bound = np.asarray(bounds).T
    diff = np.fabs(lower_bound - upper_bound)

    mu = lower_bound + diff * np.random.rand(dimensions)

    mu_value = fobj(mu)

    results = []
    all = []
    results.append((np.copy(mu), mu_value))
    sigma = sigma_init

    for i in range(max_iters):
        epsilon = np.random.randn(lambd, dimensions)
        new_solutions = mu + sigma * epsilon
        new_solutions = np.clip(new_solutions, lower_bound, upper_bound)
        new_values = np.asarray([fobj(new_solutions[i]) for i in range(lambd)])

        best_index = new_values.argmin()
        best_value = new_values[best_index]
        best_solution = new_solutions[best_index]

        if best_value <= mu_value:
            mu = best_solution.copy()
            mu_value = best_value
            sigma *= c_inc
        else:
            sigma *= c_dec

        results.append((np.copy(mu), mu_value))
        all.append(np.copy(new_solutions))
    
    return results, all
```
